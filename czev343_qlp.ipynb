```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from astropy.io import fits
```

```python
hdul = fits.open('hlsp_qlp_tess_ffi_s0019-0000000239808300_tess_v01_llc.fits')
hdul.readall()
hdul.info()
```

```python
data = hdul[1].data
time = data['TIME']
phase = time/1.20937059 - np.floor(time/1.20937059)
sap_flux = data['SAP_FLUX'] #Normalized simple aperture photometry light curve from the best aperture
kspsap_flux = data['KSPSAP_FLUX'] #Normalized light curve detrended by kepler spline
kspsap_flux_sml = data['KSPSAP_FLUX_SML'] #Detrended normalized simple aperture photometry light curve from a smaller aperture
kspsap_flux_lag = data['KSPSAP_FLUX_LAG'] #Detrended normalized simple aperture photometry light curve from a larger aperture
```

```python
plt.figure(figsize=(15,8))
plt.scatter(phase, sap_flux, s=2)
```

```python
plt.figure(figsize=(15,8))
plt.scatter(phase, kspsap_flux, s=2)
```

```python
plt.figure(figsize=(15,8))
plt.scatter(phase, kspsap_flux_sml, s=2)
```

```python
plt.figure(figsize=(15,8))
plt.scatter(phase, kspsap_flux_lag, s=2)
```
